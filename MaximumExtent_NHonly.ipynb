```python
# Import necessary packages
import pandas as pd
import numpy as np
from scipy import interpolate
from scipy.stats import linregress
from scipy import optimize
import matplotlib.pyplot as plt

# Open Northern Hemisphere data file
NHfilepath = 'data/NHAnnualExtent_Min_Max_Dates.csv'
NHmin_max = pd.read_csv(NHfilepath)
display(NHmin_max)
```

```python
# Convert min values of date columns to datetime format
NHmin_max['max-date'] = pd.to_datetime(NHmin_max['max-date'])

# filter for only 1979-1999 and 2000-2020
NH_1979 = NHmin_max[(NHmin_max['Year'] >= 1979) & (NHmin_max['Year'] <= 1999)]
NH_2000 = NHmin_max[(NHmin_max['Year'] >= 2000) & (NHmin_max['Year'] <= 2022)]

# Create 2 column subplots to compare min vs. max
fig, axs = plt.subplots(figsize=(14,8), ncols =2, sharey=True)

# set figure title 
fig.suptitle('Maximum Sea Ice Extent')
# share y axis label
fig.text(0.02, 0.5, 'Sea Ice Extent (millions of sq km)', va='center', rotation='vertical')
# share x axis label
fig.text(0.5, 0.02, 'Year', ha='center')

# first (left) subplot of min values
axs[0].plot(NH_1979['max-date'], NH_1979['max-extent'], color='teal', label='1979-1999', marker='o', markerfacecolor='palegreen')
# add figure elements
axs[0].grid()
axs[0].legend()

# second subplot (right) of NH and SH max values
axs[1].plot(NH_2000['max-date'], NH_2000['max-extent'], c='darkviolet', label='2000-2022', marker='o', markerfacecolor='lightsteelblue')
# add figure elements
axs[1].legend()
axs[1].grid()

# change fontsize
plt.rcParams.update({'font.size':20})

# save figure 
plt.savefig('MaxExtent.png')

plt.show()
```
