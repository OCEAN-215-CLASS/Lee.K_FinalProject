```python
# Import necessary packages
import pandas as pd

# Open file
filepath = 'data/N_Sea_Ice_Index_Regional_Monthly_Data_CArctic.csv'
monthly_extent = pd.read_csv(filepath)

# Get the mean of each year
monthly_extent.set_index('Year', inplace=True)
yearly_means = monthly_extent.mean(axis=1)
display(yearly_means)
```

```python
# Import NumPy, Matplotlib
import numpy as np
import matplotlib.pyplot as plt

# create plot 
plt.figure(figsize=(12,6))
# filter data for only 1979 to 1999
past_years = yearly_means.loc[1979:1999]
# filter data for only 2000-2020
now = yearly_means.loc[2000:2022]
# plot graph with two lines for the two ranges
plt.plot(past_years.index, past_years, color='teal', label='1979-1999', marker='o', markerfacecolor='palegreen')
plt.plot(now.index, now, color='darkviolet', label='2000-2022', marker='o', markerfacecolor='lightsteelblue')
# add x label
plt.xlabel('Year')
# add y label
plt.ylabel('Mean Area of Sea Ice (million sq km)')
# add title
plt.title('Monthly Mean of Sea Ice From 1979-1999 and 2000-2022')
# add legend
plt.legend()
# add grid
plt.grid()
# increase fontsize for everything 
plt.rcParams.update({'font.size':20})

# save figure
plt.savefig('MonthlyMean.png')
```
